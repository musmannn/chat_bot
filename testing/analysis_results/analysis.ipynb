```python
import pandas as pd
import json
```

```python


def json_to_dataframe(file1_path, file2_path):
    # Read JSON data from the files
    with open(file1_path, 'r') as file1, open(file2_path, 'r') as file2:
        data1 = json.load(file1)
        data2 = json.load(file2)
    
    # Combine the data into a list of dictionaries
    combined_data = []
    for i, (entry1, entry2) in enumerate(zip(data1, data2), start=1):
        combined_entry = {
            'number': i,
            'model': entry1['model'],
            'question': entry1['user_input'],
            'response_simple': entry1['response'],
            'response_example': entry2['response'],
            'tokens_simple':entry1['tokens_used'],
            'tokens_example':entry2['tokens_used']
        }
        combined_data.append(combined_entry)
    
    # Convert the combined data into a DataFrame
    df = pd.DataFrame(combined_data)
    
    return df


    # Replace 'file1.json' and 'file2.json' with the paths to your JSON files
file1_path = 'data_simple.json'
file2_path = 'data_example.json'
dataframe = json_to_dataframe(file1_path, file2_path)
data = dataframe[dataframe['model']=="mistralai/Mistral-7B-Instruct-v0.2"]
data.sort_values(by="tokens_simple",ascending=False).head(10)


dataframe.head()
csv_output_path = 'combined_data_new.csv'
dataframe.to_csv(csv_output_path)


```
